```python
#素数の判定を行う関数を定義
#素数でない場合には、リストに公約数を追加
def judge_prime(num): 
    list_s = []
    flag =0
    for x in range(2,num):
        mod = num%x
        if mod == 0:
            flag = 1
            list_s.append(x)
    if flag == 0:
        return"prime"
    else:
        return"not prime",list_s

    
#===========================================================
#プログラム本文
list_prime = []
num = int(input())
for x in range(1,num):
    if judge_prime(x)=="prime":
        list_prime.append(x)

if len(list_prime) > 1:
    print(num,"は素数ではありません")
print(list_prime)
```
